```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Scatter Plot Generation

file_scatter = "/content/drive/MyDrive/DS 200/Data Visualization/Scatter Plot/Unemployment_Participation_Per_1000_Persons_typewise_upto2012.csv"
scatter_data = pd.read_csv(file_scatter)
```

```python
x_data = scatter_data.loc[scatter_data["Particulars"] == "Current Weekly Status - Rural - Male"]
y_data = scatter_data.loc[scatter_data["Particulars"] == "Current Weekly Status - Rural - Female"]
```

```python
x = x_data.to_numpy()
x = x[0][1:]

y = y_data.to_numpy()
y = y[0][1:]
```

```python
scatter_fig = plt.figure(figsize=[5,5])

plt.scatter(x, y)
#plt.xticks(range(30,45,1))
#plt.yticks(range(30,45,1))
plt.xlabel("Rural Males")
plt.ylabel("Rural Females")
plt.grid()
plt.xlim(30, 45)
plt.ylim(30,45)
plt.title('Current Rural Unemployment Participation per 1000 Persons')
```

```python
# Line Plot Generation

file_line = "/content/drive/MyDrive/DS 200/Data Visualization/Line Plot/Statewise_General_Index_july2019_20Aug2020_dec20_2_1.csv"
line_data = pd.read_csv(file_line)
```
